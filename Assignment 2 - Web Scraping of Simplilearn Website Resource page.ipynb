```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://www.simplilearn.com/resources'
```

```python
result = requests.get(url)
page_content = result.content
soup = BeautifulSoup(page_content,'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.head
```

```python
soup.title
```

```python
for tag in soup.find_all(True):
    if tag.name == 'link' or tag.name == 'a':
        print(tag.get('href'))
```

```python
for tag in soup.find_all(True):
    if tag.name == 'h1' or tag.name == 'h2' or tag.name == 'h3' or tag.name == 'h4' or tag.name == 'h5' or tag.name == 'h6':
        print(tag)
```

```python
for tag in soup.find_all(True):
    if tag.name == 'h3':
        print(tag.get_text())
```

```python
for tag in soup.find_all(True):
    if tag.name == 'h4':
        print(tag.get_text())
```
