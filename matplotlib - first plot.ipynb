```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
randomNumber = np.random.rand(10)
```

```python
print(randomNumber)
```

```python
style.use('ggplot')
plt.plot(randomNumber,'g',label='line 1', linewidth='2')
plt.xlabel('range')
plt.ylabel('numbers')
plt.title('first plot')
plt.legend()
plt.show()
```
